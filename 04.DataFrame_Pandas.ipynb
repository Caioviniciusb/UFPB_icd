## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

```python
import pandas as pd
movies_df = pd.read_csv("../../caiov/ICD/movies.csv")
```

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
# ADICIONE O SEU CÓDIGO AQUI

movies_genre = movies_df.groupby("Major Genre").mean()    #Média dos valores dos gêneros
movies_genre.sort_values(by="Worldwide Gross", ascending=False)[["Worldwide Gross"]]
```

### Qual o diretor dirigiu mais filmes?

```python
# ADICIONE O SEU CÓDIGO AQUI

director_count = movies_df.groupby("Director").count()
director_count = director_count.sort_values(by="Title", ascending=False)[["Title"]]

director_count.iloc[0:10]
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
# ADICIONE O SEU CÓDIGO AQUI

movies_df["Profit %"] = (((movies_df["Worldwide Gross"]*100)/movies_df["Production Budget"]) - 100)
movie_gross = movies_df.sort_values(by="Profit Percentage", ascending=False)[["Title","Worldwide Gross","Production Budget","Profit %"]]

movie_gross["Worldwide Gross"] = "$" + (movie_gross["Worldwide Gross"].astype(float)/1000000).astype(str) + "M"
movie_gross["Production Budget"] = "$" + (movie_gross["Production Budget"].astype(float)/1000000).astype(str) + "M"
movie_gross["Profit %"] = (movie_gross["Profit %"].astype(float)/1000).astype(str) + "%"

movie_gross.iloc[0:10]
```
