## Medidas de Centralidade e Dispersão

A partir do dataset `movies.csv` calcule a média, mediana e desvio padrão para as colunas:

```python
import pandas as pd
pd.options.display.float_format = "{:.3f}".format

movies_df = pd.read_csv("../../caiov/ICD/movies.csv")
movies_df
```

### US Gross

```python
# ADICIONE O SEU CÓDIGO AQUI

Gross = {"Média": movies_df["US Gross"].mean(),
                  "Mediana": movies_df["US Gross"].median(),
                  "Desvio Padrão": movies_df["US Gross"].std()}
gross = pd.Series(Gross)

Gross_Values = pd.DataFrame({"US Gross": gross})
Gross_Values
```

### Production Budget

```python
# ADICIONE O SEU CÓDIGO AQUI

PB = {"Média": movies_df["Production Budget"].mean(),
                  "Mediana": movies_df["Production Budget"].median(),
                  "Desvio Padrão": movies_df["Production Budget"].std()}
pb = pd.Series(PB)

PB_Values = pd.DataFrame({"Production Budget": pb})
PB_Values
```

### Running Time

```python
# ADICIONE O SEU CÓDIGO AQUI

RT = {"Média": movies_df["Running Time (min)"].mean(),
                  "Mediana": movies_df["Running Time (min)"].median(),
                  "Desvio Padrão": movies_df["Running Time (min)"].std()}
rt = pd.Series(RT)

RT_Values = pd.DataFrame({"Running Time": rt})
RT_Values
```

### IMDB Rating

```python
# ADICIONE O SEU CÓDIGO AQUI

IMDB_R = {"Média": movies_df["IMDB Rating"].mean(),
                  "Mediana": movies_df["IMDB Rating"].median(),
                  "Desvio Padrão": movies_df["IMDB Rating"].std()}
imdb_r = pd.Series(IMDB_R)

IMDB_R_Values = pd.DataFrame({"IMDB Rating": imdb_r})
IMDB_R_Values
```

### IMDB Votes

```python
# ADICIONE O SEU CÓDIGO AQUI

IMDB_V = {"Média": movies_df["IMDB Votes"].mean(),
                  "Mediana": movies_df["IMDB Votes"].median(),
                  "Desvio Padrão": movies_df["IMDB Votes"].std()}
imdb_v = pd.Series(IMDB_V)

IMDB_V_Values = pd.DataFrame({"IMDB Votes": imdb_v})
IMDB_V_Values
```

### Análise dos Resultados

Quais colunas tiveram valores muito diferentes entre média e mediana? Quais os seus desvios padrão?

**As colunas US Gross (62555311.391 - DV), Production Budget (35585913.445 - DV) e a IMDB Votes (44937.582 - DV).** 

Que conclusões podemos tirar desses resultados?

**Que os valores dessa coluna em relação a sua média são um pouco distantes um do outro, resultando em um alto valor de desvio padrão. Mostra também que provavelmente os valores mínimo e máximo de cada coluna também são distantes um do outro. Além disso caso não tivéssemos o desvio padrão neste Data Frame provavelmente o entendimento apenas com a média seria prejudicado, já que esta não revela a realidade dos valores da coluna.**
